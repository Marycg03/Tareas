## Ejercicio 8: Obtener todos los valores
Imprime todos los valores del diccionario fruits = {"apple": 3, "banana": 5, "orange": 2}

## Ejercicio 9: Iterar sobre claves y valores
Usa un bucle para imprimir cada clave y su valor correspondiente del diccionario fruits = {"apple": 3, "banana": 5, "orange": 2}

```python
fruits = {"apple": 3, "banana": 5, "orange": 2}
for key, value in fruits.items():
    print(f"{key}: {value}")
```

## Ejercicio 14: Iterar sobre países y capitales
Usa un bucle para imprimir cada país y su capital del diccionario countries = {"Colombia": "Bogota", "France": "Paris", "Japan": "Tokyo"}

```python
countries = {"Colombia": "Bogota", "France": "Paris", "Japan": "Tokyo"}
for country, capital in countries.items():
    print(f"{country}: {capital}")
```
