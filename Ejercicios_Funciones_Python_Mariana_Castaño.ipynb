## Ejercicio 3: Función que retorna la suma
Escribe una función llamada add(a, b) que reciba dos números y retorne su suma. Imprime el resultado de llamar la función con dos valores.

```python
def add(a, b):
    return a + b

resultado = add(5, 3)
print(resultado)
```

## Ejercicio 4: Función que retorna la diferencia

```python
def subtract(a, b):
    return a - b

resultado = subtract(10, 3)
print(resultado)
```

## Ejercicio 5: Función que retorna el producto

```python
def multiply(a, b):
    return a * b

resultado = multiply(4, 5)
print(resultado)
```

## Ejercicio 6: Función que divide dos números
Escribe una función llamada divide(a, b) que reciba dos números y retorne el resultado de dividir a entre b.

```python
def divide(a, b):
    if b != 0:
        return a / b
    else:
        return "No se puede dividir entre cero"

resultado = divide(20, 4)
print(resultado)
```

## Ejercicio 7: Función que verifica si un número es par
Escribe una función llamada is_even(number) que retorne True si un número es par y False de lo contrario.

```python
def is_even(number):
    return number % 2 == 0

print(is_even(4))
print(is_even(7))
```

## Ejercicio 8: Función que encuentra el número mayor
Escribe una función llamada find_max(a, b) que retorne el mayor de dos números.

```python
def find_max(a, b):
    return max(a, b)

resultado = find_max(15, 8)
print(resultado)
```

## Ejercicio 11: Función que cuenta las letras

## Ejercicio 14: Función que calcula el promedio
Escribe una función llamada average(numbers) que reciba una lista de números y retorne el promedio.

```python
def average(numbers):
    return sum(numbers) / len(numbers)

resultado = average([10, 20, 30, 40])
print(resultado)
```

## Ejercicio 15: Función que calcula el factorial
Escribe una función llamada factorial(n) que retorne el factorial de un entero positivo n.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

resultado = factorial(5)
print(resultado)
```
