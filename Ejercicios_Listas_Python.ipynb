# Ejercicios de Listas en Python


## Ejercicio 1: Crear una lista de frutas
Crea una lista llamada 'frutas' que contenga: "apple", "banana", "orange" y "grape". Imprime la lista.

```python
frutas = ["apple", "banana", "orange", "grape"]
print(frutas)
```

## Ejercicio 2: Primer elemento
Imprime el primer elemento de la lista numbers = [4, 8, 15, 16, 23, 42]

```python
numbers = [4, 8, 15, 16, 23, 42]
print(numbers[0])
```

## Ejercicio 3: Último elemento
Imprime el último elemento de la lista numbers = [4, 8, 15, 16, 23, 42]

```python
numbers = [4, 8, 15, 16, 23, 42]
print(numbers[-1])
```

## Ejercicio 4: Segundo y tercer elemento
Imprime el segundo y tercer elemento de la lista colors = ["red", "blue", "green", "yellow"]

```python
colors = ["red", "blue", "green", "yellow"]
print(colors[1], colors[2])
```

## Ejercicio 5: Número de elementos
Crea una lista llamada 'ciudades' con cinco nombres de ciudades. Imprime el número de elementos.

```python
ciudades = ["Madrid", "Barcelona", "Valencia", "Sevilla", "Bilbao"]
print(len(ciudades))
```

## Ejercicio 6: Añadir elemento al final
Añade "hamster" al final de la lista animals = ["dog", "cat", "rabbit"] e imprime la lista actualizada.

```python
animals = ["dog", "cat", "rabbit"]
animals.append("hamster")
print(animals)
```

## Ejercicio 7: Insertar elemento en posición
Inserta "bird" en la segunda posición de la lista animals = ["dog", "cat", "rabbit"] e imprime la lista actualizada.

```python
animals = ["dog", "cat", "rabbit"]
animals.insert(1, "bird")
print(animals)
```

## Ejercicio 8: Eliminar un valor
Elimina el valor 30 de la lista numbers = [10, 20, 30, 40, 50] e imprime la lista actualizada.

```python
numbers = [10, 20, 30, 40, 50]
numbers.remove(30)
print(numbers)
```

## Ejercicio 9: Eliminar último elemento con pop()
Elimina el último elemento de la lista letters = ["a", "b", "c", "d", "e"] usando pop() e imprime la lista actualizada.

```python
letters = ["a", "b", "c", "d", "e"]
letters.pop()
print(letters)
```

## Ejercicio 10: Verificar si un elemento está en la lista
Verifica si "Maria" está en la lista names = ["Ana", "Luis", "Maria", "Pedro"]. Imprime True o False.

```python
names = ["Ana", "Luis", "Maria", "Pedro"]
print("Maria" in names)
```

## Ejercicio 11: Ordenar en orden ascendente
Ordena la lista numbers = [5, 2, 9, 1, 7] en orden ascendente e imprime el resultado.

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort()
print(numbers)
```

## Ejercicio 12: Ordenar en orden descendente
Ordena la lista numbers = [5, 2, 9, 1, 7] en orden descendente e imprime el resultado.

```python
numbers = [5, 2, 9, 1, 7]
numbers.sort(reverse=True)
print(numbers)
```

## Ejercicio 13: Crear nueva lista con elementos multiplicados
Crea una nueva lista con cada número multiplicado por 2 de la lista numbers = [3, 6, 9, 12, 15].

```python
numbers = [3, 6, 9, 12, 15]
nueva_lista = [x * 2 for x in numbers]
print(nueva_lista)
```

## Ejercicio 14: Iterar sobre una lista con un bucle
Usa un bucle para imprimir cada palabra en una línea separada de la lista words = ["python", "list", "exercise", "practice"].

```python
words = ["python", "list", "exercise", "practice"]
for word in words:
    print(word)
```

## Ejercicio 15: Calcular el promedio
Calcula e imprime el promedio de las calificaciones de la lista grades = [80, 75, 90, 85, 95].

```python
grades = [80, 75, 90, 85, 95]
promedio = sum(grades) / len(grades)
print(promedio)
```
